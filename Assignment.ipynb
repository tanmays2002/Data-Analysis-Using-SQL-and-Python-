```python
#Problem 1: Find the total number of crimes recorded in the CRIME table.
cursor.execute("""select count(*) from CHICAGO_CRIME_DATA;
                    """)
result=cursor.fetchall()
result
```

```python
#Problem 2: List community areas with per capita income less than 11000.
cursor.execute("""select COMMUNITY_AREA_NAME, PER_CAPITA_INCOME
                    from CENSUS_DATA
                    where PER_CAPITA_INCOME<11000;
                    """)
result=cursor.fetchall()
result
```

```python
#Problem 4: List all kidnapping crimes involving a child?
#(children are not considered minors for the purposes of crime analysis)
cursor.execute("""select CASE_NUMBER
                    from CENSUS_DATA
                    where PER_CAPITA_INCOME<11000;
                    """)
result=cursor.fetchall()
result
```

```python
#Problem 5: What kind of crimes were recorded at schools?
cursor.execute("""select PRIMARY_TYPE
                    from CHICAGO_CRIME_DATA
                    where LOCATION_DESCRIPTION in ('SCHOOL, PUBLIC, BUILDING','COLLEGE/UNIVERSITY GROUNDS');
                    """)
result=cursor.fetchall()
result
```

```python
#Problem 6: List the average safety score for all types of schools.
cursor.execute("""select NAME_OF_SCHOOL,avg(SAFETY_SCORE)
                    from CHICAGO_PUBLIC_SCHOOLS
                    group by NAME_OF_SCHOOL;
                    """)
result=cursor.fetchall()
result
```

```python
#Problem 7: List 5 community areas with highest % of households below poverty line.
cursor.execute("""select COMMUNITY_AREA_NAME,PERCENT_HOUSEHOLDS_BELOW_POVERTY
                    from CENSUS_DATA
                    order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc
                    limit 5;
                    """)
result=cursor.fetchall()
result
```

```python
#Problem 9: Use a sub-query to find the name of the community area with highest hardship index.
cursor.execute("""select COMMUNITY_AREA_NAME,HARDSHIP_INDEX
                    from CENSUS_DATA
                    order by HARDSHIP_INDEX desc
                    limit 1;
                    """)
result=cursor.fetchall()
result
```
